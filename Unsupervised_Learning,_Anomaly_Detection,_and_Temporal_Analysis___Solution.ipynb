```python
Question 8: Write Python code to perform K-means clustering on a sample dataset.

Program =>

# Import necessary libraries
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Create a sample dataset
X, y = make_blobs(n_samples=200, centers=3, random_state=42, cluster_std=1.2)

# Step 2: Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Step 3: Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Step 4: Visualize the clusters
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X')
plt.title("K-Means Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Step 5: Print cluster centroids
print("Cluster Centroids:")
print(centroids)
```

```python
Question 9: What is inheritance in OOP? Provide a simple example in Python.

Answer:Inheritance in Object-Oriented Programming (OOP) is a concept where one
class (called the child or subclass) can inherit attributes and methods from
another class (called the parent or superclass).

It allows code reusability and makes it easier to extend functionality without
rewriting existing code.

Program =>

# Parent class
class Animal:
    def speak(self):
        return "Animal makes a sound"

# Child class (inherits from Animal)
class Dog(Animal):
    def speak(self):
        return "Dog barks"

# Creating objects
a = Animal()
d = Dog()

print(a.speak())  # Output: Animal makes a sound
print(d.speak())  # Output: Dog barks
```

Question 1 : What is Dimensionality Reduction? Why is it important in machine
learning?

Answer: Dimensionality Reduction is the process of reducing the number of input variables or features in a dataset while retaining as much relevant information as possible. In simpler terms, it means converting high-dimensional data (with many features) into a lower-dimensional form that still captures the key patterns and relationships.

Why Its Important in Machine Learning:

1. Removes Redundant and Irrelevant Features:
Many features in datasets carry overlapping or useless information. Reducing dimensions helps eliminate noise and improves model clarity.

2. Prevents Overfitting:
With fewer irrelevant features, the model generalizes better and doesnt memorize noise from the training data.

3. Improves Model Performance and Speed:
Algorithms run faster and use less memory when trained on reduced feature sets.

4. Enhances Visualization:
Its impossible to visualize data beyond 3 dimensions. Dimensionality reduction (like PCA or t-SNE) allows visualization of complex data in 2D or 3D.

5. Simplifies Models:
Fewer features often mean simpler, more interpretable models.

Common Techniques:

* Principal Component Analysis (PCA)
* Linear Discriminant Analysis (LDA)
* t-Distributed Stochastic Neighbor Embedding (t-SNE)
* Autoencoders (Neural Network-based)

Question 2: Name and briefly describe three common dimensionality reduction
techniques..

Answer: Here are three common dimensionality reduction techniques used in machine learning:


1. Principal Component Analysis (PCA):

* A linear technique that transforms the original correlated features into a smaller set of uncorrelated features called principal components.
* These components capture the maximum variance in the data.
* Its widely used for noise reduction, visualization, and speeding up model training.

2. Linear Discriminant Analysis (LDA):

* A supervised technique that reduces dimensions by maximizing the separation between multiple classes.
* It projects data onto a lower-dimensional space where different categories are as distinct as possible.
* Often used in classification problems like face recognition or document categorization.

3. t-Distributed Stochastic Neighbor Embedding (t-SNE):

* A non-linear technique mainly used for visualizing high-dimensional data in 2D or 3D.
* It preserves local structure (similar data points stay close) and is excellent for understanding data clusters.
* Commonly used in exploratory data analysis.

Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

Answer: Clustering is an unsupervised learning technique used to group similar data points into clusters based on their features  without using predefined labels.
The goal is to ensure that data points within the same cluster are highly similar to each other, while those in different clusters are as different as possible.

Example:

In customer segmentation, clustering can group customers with similar buying habits without prior knowledge of their categories.

Three Popular Clustering Algorithms:

1. K-Means Clustering:

* Divides data into k clusters based on distance from cluster centroids.
* Simple and fast, best for spherical, well-separated clusters.

2.Hierarchical Clustering:

* Builds a tree-like structure (dendrogram) of clusters by repeatedly merging or splitting clusters based on similarity.
* Doesnt require specifying the number of clusters in advance.

3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise):

* Groups together points that are close in dense regions and marks outliers as noise.
* Works well with clusters of arbitrary shape and handles noise effectively.

Question 4: Explain the concept of anomaly detection and its significance.

Answer: Anomaly Detection is the process of identifying data points, events, or patterns that deviate significantly from the normal behavior in a dataset.
These unusual instances are called anomalies or outliers and often indicate critical information such as fraud, system failure, or errors.

Significance:

1. Fraud Detection:

* Helps identify unusual transactions in banking, credit cards, or online payments that may indicate fraud.

2. System Health Monitoring:

* Detects abnormal behavior in machines, servers, or networks before a breakdown occurs.

3. Security Threat Detection:

* Identifies unauthorized access or suspicious activities in cybersecurity systems.

4. Data Quality Improvement:

* Removes incorrect or inconsistent records that can distort analysis and degrade model performance.

5. Early Warning Systems:

* Useful in healthcare or finance for spotting early signs of risk or failure.

Question 5: List and briefly describe three types of anomaly detection techniques.

Answer: Here are three main types of anomaly detection techniques used in machine learning:

1. Statistical (Model-Based) Methods:

* These methods assume data follows a certain statistical distribution (e.g., normal distribution).
* Any data point that lies far from the expected range (like beyond 3 standard deviations from the mean) is marked as an anomaly.
* Example: Z-score, Gaussian models.

2. Distance-Based Methods:

* These techniques measure how far each point is from others in the dataset.
* Points that are far away from most others are considered anomalies.
* Example: K-Nearest Neighbors (KNN)-based anomaly detection, Euclidean distance methods.

3. Density-Based Methods:

* Focus on the density of data points in an area  anomalies appear in low-density regions.
* Example: DBSCAN, Local Outlier Factor (LOF).

Question 6: What is time series analysis? Mention two key components of time series
data.

Answer: Time Series Analysis is a statistical and machine learning technique used to analyze data points collected or recorded over time  usually at regular intervals (e.g., hourly, daily, monthly).
The main goal is to identify patterns, trends, and relationships in the data to make forecasts or predictions about future values.

Example:

Stock prices, weather data, electricity usage, and sales data are all common examples of time series.

Two Key Components of Time Series Data:

1. Trend:

* The long-term direction or movement of data over time.
* It can be upward (growth), downward (decline), or flat (no major change).

2. Seasonality:

* Repeating patterns or fluctuations that occur at regular intervals (e.g., daily, monthly, or yearly).
* Example: Higher ice cream sales in summer and lower in winter.

Question 7: Describe the difference between seasonality and cyclic behavior in time
series.

Answer: Both seasonality and cyclic behavior refer to repeating patterns in time series data, but they differ in duration and predictability.

1. Seasonality:

* Refers to regular and predictable patterns that repeat at fixed intervals (such as daily, monthly, or yearly).
* These patterns are usually driven by calendar or time-based factors.
* Example:
1.Ice cream sales increasing every summer.
2. Electricity usage peaking every evening.

Key Point: Seasonal effects occur at consistent, known periods.

2. Cyclic Behavior:

* Refers to irregular and unpredictable fluctuations that occur over longer and varying time spans.
* Usually influenced by economic or natural cycles, not fixed dates.
* Example:
1. Business boom and recession cycles.
2. Long-term price fluctuations in housing or stock markets.

Key Point: Cyclic effects have no fixed frequency  they depend on external factors and can last for months or years.

Question 10: How can time series analysis be used for anomaly detection?

Answer: Time Series Analysis for Anomaly Detection involves identifying unusual patterns or data points that deviate significantly from the expected time-based behavior.
Since time series data is ordered chronologically, detecting anomalies means spotting points that break the normal trend, seasonality, or expected range.

How Its Done:


1. Modeling Normal Behavior:

* First, a time series model (like ARIMA, Exponential Smoothing, or LSTM) is trained to learn normal patterns such as trends and seasonality.

2. Predicting Expected Values:

* The model forecasts the expected value at each time step.

3. Detecting Deviations:

* Actual observations are compared with predicted values.
* If the difference (error or residual) exceeds a certain threshold, that point is flagged as an anomaly.

Example Applications:

* Finance: Detecting fraudulent transactions or sudden market drops.
* IT Systems: Spotting abnormal spikes in server load or latency.
* Manufacturing: Identifying equipment failures from sensor data.
